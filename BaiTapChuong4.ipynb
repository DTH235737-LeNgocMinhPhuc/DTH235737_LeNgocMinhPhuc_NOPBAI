```python
#câu 7
import math

def calculate_distance():
    """Nhập tọa độ 2 điểm và tính khoảng cách Euclid giữa chúng."""
    print("--- Nhập tọa độ điểm A ---")
    try:
        # Nhập tọa độ xA, yA
        xA = float(input("Nhập xA: "))
        yA = float(input("Nhập yA: "))

        print("\n--- Nhập tọa độ điểm B ---")
        # Nhập tọa độ xB, yB
        xB = float(input("Nhập xB: "))
        yB = float(input("Nhập yB: "))

        # Tính toán theo công thức
        delta_x = xB - xA
        delta_y = yB - yA

        # Công thức: căn bậc hai của (delta_x^2 + delta_y^2)
        distance_sq = delta_x**2 + delta_y**2
        distance = math.sqrt(distance_sq)

        # Xuất kết quả
        print(f"\n Tọa độ A: ({xA}, {yA})")
        print(f" Tọa độ B: ({xB}, {yB})")
        print(f"**Độ dài đoạn AB (|AB|) là: {distance:.4f}**") # Làm tròn 4 chữ số thập phân

    except ValueError:
        print("\n Lỗi: Vui lòng nhập giá trị số hợp lệ cho tọa độ.")

# Chạy hàm
calculate_distance()
```

```python
#Câu 8
import math

def calculate_log_base_a():
    """Nhập x, a và tính log_a x."""
    print("--- Tính giá trị log_a x ---")
    try:
        # Nhập giá trị x và a
        x = float(input("Nhập giá trị x (x > 0): "))
        a = float(input("Nhập cơ số a (a > 0 và a != 1): "))

        # 1. Kiểm tra điều kiện tồn tại của logarit
        if x <= 0:
            print("\n Lỗi: Giá trị x phải lớn hơn 0.")
            return

        if a <= 0:
            print("\n Lỗi: Cơ số a phải lớn hơn 0.")
            return

        if a == 1:
            print("\n Lỗi: Cơ số a phải khác 1.")
            return

        # 2. Áp dụng công thức chuyển đổi cơ số: log_a x = ln(x) / ln(a)
        # math.log() trong Python tính logarit tự nhiên (ln)
        log_x = math.log(x)
        log_a = math.log(a)

        result = log_x / log_a

        # 3. Xuất kết quả
        print(f"\n Với x = {x} và a = {a}:")
        print(f"**log({a})({x}) = {result:.6f}**") # Làm tròn 6 chữ số thập phân

    except ValueError:
        print("\n Lỗi nhập liệu: Vui lòng nhập giá trị số thực hợp lệ.")

# Chạy hàm
calculate_log_base_a()
```

```python
# Câu 9
import math

def calculate_nested_square_roots():
    """Nhập n và tính giá trị S(n) của căn bậc hai lồng nhau."""
    print("--- Tính giá trị S(n) = sqrt(2 + sqrt(2 + ... + sqrt(2))) ---")
    try:
        # Nhập số lần lồng nhau n
        n = int(input("Nhập số lần lồng nhau (n là số nguyên dương): "))

        if n <= 0:
            print("\n Lỗi: n phải là số nguyên dương.")
            return

        # Khởi tạo giá trị S ban đầu (từ trong cùng)
        # S_0 = sqrt(2)
        S = math.sqrt(2)

        # Bắt đầu vòng lặp từ i = 2 đến n
        # Nếu n=1, vòng lặp không chạy, S vẫn là sqrt(2), đúng.
        # Nếu n=2, vòng lặp chạy 1 lần (i=2), tính sqrt(2 + sqrt(2)).
        for i in range(2, n + 1):
            # Công thức lặp: S_i = sqrt(2 + S_(i-1))
            S = math.sqrt(2 + S)

        # Xuất kết quả
        print(f"\nVới n = {n} lần lồng nhau, giá trị S(n) là:")
        print(f"**S({n}) = {S:.8f}**") # Làm tròn 8 chữ số thập phân

    except ValueError:
        print("\n Lỗi nhập liệu: Vui lòng nhập n là số nguyên hợp lệ.")

# Chạy hàm
calculate_nested_square_roots()
```

```python
#Câu 12
def oscillate(start, end):

    for i in range(abs(start), 0, -1):
        yield -i
        yield i
    for i in range(end):
        if i == 0:
            yield 0
            yield 0
            continue
        yield i
        yield -i

result = []
for n in oscillate(-3, 5):
    result.append(str(n))

print("Kết quả mong muốn:")
print("-3 3 -2 2 -1 1 0 0 1 -1 2 -2 3 -3 4 -4")

print("\nKết quả thực hiện:")
print(" ".join(result))
```

```python
def sum_proper_divisors(n):
    """
    Tính tổng các ước số thực sự của số nguyên dương n
    (các ước số nhỏ hơn n, không tính n).
    """
    if n <= 0:
        return 0
    if n == 1:
        return 0
    sum_divisors = 1
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            sum_divisors += i
            j = n // i
            if i != j:
                sum_divisors += j

    return sum_divisors



def is_perfect_number(n):
    """
    a) Kiểm tra số hoàn thiện (Perfect number):
    Tổng các ước số thực sự bằng chính nó.
    """
    if n <= 1:
        return False

    return sum_proper_divisors(n) == n


def is_abundant_number(n):
    """
    b) Kiểm tra số thịnh vượng (Abundant number):
    Tổng các ước số thực sự lớn hơn chính nó.
    """
    if n <= 1:
        return False

    return sum_proper_divisors(n) > n

try:
    N = int(input("Nhập một số nguyên dương n để kiểm tra: "))

    if N <= 0:
        print("Vui lòng nhập số nguyên dương.")
    else:
        s = sum_proper_divisors(N)
        print(f"\n--- Phân tích số {N} ---")
        print(f"Tổng các ước số thực sự (không kể {N}): **{s}**")

        # Kiểm tra Số hoàn thiện
        if is_perfect_number(N):
            print(f" Số {N} là **Số Hoàn Thiện** (vì {s} = {N}).")
        else:
            print(f" Số {N} **không** là Số Hoàn Thiện.")

        # Kiểm tra Số thịnh vượng
        if is_abundant_number(N):
            print(f" Số {N} là **Số Thịnh Vượng** (vì {s} > {N}).")
        else:
            print(f" Số {N} **không** là Số Thịnh Vượng.")

        # Kiểm tra thêm số Khuyết thiếu (Deficient number)
        if s < N:
            print(f" Số {N} là **Số Khuyết Thiếu** (vì {s} < {N}).")

except ValueError:
    print(" Lỗi nhập liệu: Vui lòng nhập một số nguyên hợp lệ.")
```
