```python
#Câu 4
"""
Nhóm Kiểm Tra Trạng Thái (Boolean Methods):
Các hàm này được dùng để kiểm tra nội dung của chuỗi và trả về giá trị True hoặc False.
.isalpha(): Kiểm tra xem tất cả các ký tự trong chuỗi có phải là chữ cái (A-z) hay không.
.isdigit(): Kiểm tra xem chuỗi có chỉ chứa chữ số (0-9) hay không.
.isalnum(): Kiểm tra xem chuỗi có chỉ chứa chữ cái hoặc chữ số hay không.
.isspace(): Kiểm tra xem chuỗi có chỉ chứa khoảng trắng hay không.
.startswith(sub): Kiểm tra xem chuỗi có bắt đầu bằng chuỗi con sub được chỉ định hay không.
.endswith(sub): Kiểm tra xem chuỗi có kết thúc bằng chuỗi con sub được chỉ định hay không.

Nhóm Thay Đổi Định Dạng (Formatting Methods):
Các hàm này trả về một chuỗi mới đã được định dạng (chuỗi gốc không thay đổi vì chuỗi là bất biến).
.upper(): Chuyển tất cả các ký tự sang chữ hoa.
.lower(): Chuyển tất cả các ký tự sang chữ thường.
.strip(): Xóa khoảng trắng (hoặc ký tự chỉ định) ở đầu và cuối chuỗi. Đây là hàm rất quan trọng khi xử lý dữ liệu nhập từ người dùng.
.replace(old, new): Thay thế tất cả các lần xuất hiện của chuỗi con $old$ bằng chuỗi new

 Nhóm Tìm Kiếm và Tách Chuỗi (Search & Split Methods):
 .find(sub): Trả về chỉ số (vị trí) đầu tiên mà chuỗi con sub xuất hiện. Nếu không tìm thấy, nó trả về -1.
 .count(sub): Đếm số lần chuỗi con $sub$ xuất hiện trong chuỗi.
 .split(sep): Tách chuỗi thành một list các chuỗi con, dựa trên dấu phân cách $sep$ (nếu sep không được chỉ định, nó sẽ tách bằng khoảng trắng).
 .join(iterable): Đây là một hàm đảo ngược của .split(). Nó nối các phần tử trong một iterable thành một chuỗi duy nhất, sử dụng chuỗi gốc (string gọi hàm) làm dấu phân cách giữa các phần tử.
 """
```

```python
#Câu 5
def analyze_string_input():
    """Nhập chuỗi từ bàn phím và phân tích các loại ký tự."""

    # Khởi tạo bộ đếm
    count_upper = 0
    count_lower = 0
    count_digit = 0
    count_space = 0
    count_vowel = 0
    count_consonant = 0

    # Nhập chuỗi
    s = input("Vui lòng nhập một chuỗi bất kỳ: ")

    # Định nghĩa tập hợp nguyên âm (chỉ cần chữ thường)
    vowels = "aeiouáàảãạăắằẳẵặâấầẩẫậéèẻẽẹêếềểễệíìỉĩịóòỏõọôốồổỗộơớờởỡợúùủũụưứừửữựýỳỷỹỵ"

    for char in s:
        # 1. Chữ hoa, chữ thường, chữ số, khoảng trắng
        if char.isupper():
            count_upper += 1
        elif char.islower():
            count_lower += 1
        elif char.isdigit():
            count_digit += 1
        elif char.isspace():
            count_space += 1

        # 2. Nguyên âm và Phụ âm (chỉ xét các chữ cái)
        if char.isalpha():
            char_lower = char.lower()
            if char_lower in vowels:
                count_vowel += 1
            else:
                count_consonant += 1

    # Tính ký tự đặc biệt (tất cả các ký tự không phải chữ cái, số, hay khoảng trắng)
    count_special = len(s) - (count_upper + count_lower + count_digit + count_space)

    # Xuất kết quả
    print("\n--- KẾT QUẢ PHÂN TÍCH CHUỖI ---")
    print(f"Tổng số ký tự: {len(s)}")
    print(f"* Số chữ IN HOA: {count_upper}")
    print(f"* Số chữ in thường: {count_lower}")
    print(f"* Số chữ là chữ số: {count_digit}")
    print(f"* Số chữ là khoảng trắng: {count_space}")
    print(f"* Số ký tự đặc biệt: {count_special}")
    print(f"* Số chữ là Nguyên Âm: {count_vowel}")
    print(f"* Số chữ là Phụ Âm: {count_consonant}")

# Chạy hàm
analyze_string_input()
```

```python
#Câu 6
import re

def NegativeNumberInStrings(s):

    negative_numbers_str = re.findall(r'-\d+', s)

    if not negative_numbers_str:
        print(f"Không tìm thấy số nguyên âm nào trong chuỗi: '{s}'")
        return

    print(f"Các số nguyên âm được trích lọc từ chuỗi là:")
    for num_str in negative_numbers_str:

        num_int = int(num_str)
        print(f"**{num_int}**")

print("--- Ví dụ 1 ---")
input_str_1 = "abc-5xyz-12k9l--p"
NegativeNumberInStrings(input_str_1)
print("\n--- Ví dụ 2 ---")
input_str_2 = "0-10a-200c30-4-0"
NegativeNumberInStrings(input_str_2)
print("\n--- Ví dụ 3 (Không có số âm) ---")
input_str_3 = "123x-a456"
NegativeNumberInStrings(input_str_3)
```

```python
#Câu 7
def optimize_name_string():

    input_str = input("Vui lòng nhập chuỗi danh từ (ví dụ: tên): ")
    words = input_str.strip().split()

    if not words:
        print("❌ Chuỗi nhập vào rỗng hoặc chỉ chứa khoảng trắng.")
        return
    optimized_words = [word.capitalize() for word in words]
    optimized_str = " ".join(optimized_words)
    print("\n--- KẾT QUẢ TỐI ƯU ---")
    print(f"Input: '{input_str}'")
    print(f"Output: '{optimized_str}'")

optimize_name_string()
```

```python
#câu 8
import os

def extract_filename(path):
    """
    1. Lấy ra tên tệp tin đầy đủ (ví dụ: muabui.mp3)
    """
    # os.path.basename() trả về phần cuối cùng của đường dẫn (tên file)
    filename = os.path.basename(path)
    return filename

def extract_basename(path):
    """
    2. Lấy ra tên tệp tin không có phần mở rộng (ví dụ: muabui)
    """
    # Bước 1: Lấy tên tệp tin đầy đủ
    filename_full = extract_filename(path)

    # Bước 2: Tách tên tệp tin và phần mở rộng
    # os.path.splitext() trả về một tuple (basename, extension)
    basename, extension = os.path.splitext(filename_full)
    return basename

# --- NHẬP TỪ BÀN PHÍM ---
path_input = input("Vui lòng nhập đường dẫn file nhạc bất kỳ : ")

print("\n--- KẾT QUẢ TÁCH TÊN BÀI HÁT ---")
try:
    # Kết quả 1: Tên tệp tin đầy đủ
    filename_full = extract_filename(path_input)
    print(f"1. Tên tệp tin đầy đủ: **{filename_full}**")

    # Kết quả 2: Tên bài hát (không mở rộng)
    basename = extract_basename(path_input)
    print(f"2. Tên bài hát (không mở rộng): **{basename}**")

except Exception as e:
    print(f" Lỗi xử lý đường dẫn: {e}")
```
