```python
#Câu 2
FILE_PATH_2 = "numbers.txt"
data_to_write = """5,6,8,9,-5
-9.5,4,7,8
6,7,8,3,6,46,7.2,-6,-7
"""

try:
    with open(FILE_PATH_2, 'w', encoding='utf-8') as file:
        file.write(data_to_write)
    print(f"✅ Đã tạo file dữ liệu mẫu: {FILE_PATH_2}")
except Exception as e:
    print(f"❌ Lỗi khi tạo file: {e}")

def DocFileVaXuLy(path):
    print("\n--- Yêu cầu a) Đọc file, mỗi dòng tạo thành 1 list và xuất ra màn hình ---")
    try:
        with open(path, 'r', encoding='utf-8') as file:
            all_lists = []
            print("\n--- Yêu cầu b) Xuất các số âm trên mỗi dòng ra màn hình ---")

            for i, line in enumerate(file):
                data_line = line.strip()
                if not data_line: continue
                string_numbers = data_line.split(',')
                list_of_floats = []
                negative_numbers_in_line = []

                for num_str in string_numbers:
                    try:
                        number = float(num_str)
                        list_of_floats.append(number)
                        if number < 0:
                            negative_numbers_in_line.append(number)

                    except ValueError:
                        print(f"Cảnh báo: Bỏ qua giá trị không hợp lệ '{num_str}' ở dòng {i+1}")
                all_lists.append(list_of_floats)
                print(f"Dòng {i+1} (List): {list_of_floats}")
                if negative_numbers_in_line:
                    print(f"   -> Số âm dòng {i+1}: {negative_numbers_in_line}")
                else:
                    print(f"   -> Số âm dòng {i+1}: Không có")

            print("\n--- Kết thúc xử lý ---")
            return all_lists

    except FileNotFoundError:
        print(f" Lỗi: Không tìm thấy file '{path}'. Vui lòng đảm bảo đã chạy bước tạo file.")
        return []
    except Exception as e:
        print(f" Lỗi trong quá trình đọc và xử lý file: {e}")
        return []
DocFileVaXuLy(FILE_PATH_2)
```

```python
#Câu 7
import xlsxwriter
FILE_NAME = 'demo.xlsx'
product_data = [
    [1, 'SP1', 'Coca', 15, 15000],
    [2, 'SP2', 'Pepsi', 20, 18000],
]
try:
    workbook = xlsxwriter.Workbook(FILE_NAME)
    worksheet = workbook.add_worksheet()
    bold_format = workbook.add_format({'bold': True, 'align': 'center'})
    money_format = workbook.add_format({'num_format': '#,##0'})
    quantity_format = workbook.add_format({'align': 'center'})
    worksheet.set_column('A:A', 5, quantity_format)
    worksheet.set_column('B:B', 15)
    worksheet.set_column('C:C', 20)
    worksheet.set_column('D:D', 10, quantity_format)
    worksheet.set_column('E:E', 15, money_format)
    headings = ['STT', 'MÃ SẢN PHẨM', 'TÊN SẢN PHẨM', 'SỐ LƯỢNG', 'ĐƠN GIÁ']
    col = 0
    for header in headings:
        worksheet.write(0, col, header, bold_format)
        col += 1
    row_num = 1
    for stt, ma, ten, so_luong, don_gia in product_data:
        worksheet.write(row_num, 0, stt)
        worksheet.write(row_num, 1, ma)
        worksheet.write(row_num, 2, ten)
        worksheet.write(row_num, 3, so_luong)
        worksheet.write(row_num, 4, don_gia)
        row_num += 1
    worksheet.insert_image('B5', 'logo_UEL.png')
    workbook.close()

    print(f"\n Đã tạo file Excel thành công: {FILE_NAME}")

except Exception as e:
    print(f"\nLỗi khi xử lý Excel: {e}")
```

```python
#Câu 9
import json
import os
FILE_NAME = "product_manager_data.json"
data = []

def LuuDuLieu(file_path, du_lieu):
    """Ghi toàn bộ cấu trúc dữ liệu vào file JSON."""
    try:
        with open(file_path, 'w', encoding='utf-8') as f:
            json.dump(du_lieu, f, ensure_ascii=False, indent=4)
        print(f"Đã lưu dữ liệu thành công vào {file_path}")
    except Exception as e:
        print(f" Lỗi khi lưu file: {e}")

def DocDuLieu(file_path):
    """Đọc dữ liệu từ file JSON. Nếu file không tồn tại, trả về list rỗng."""
    if not os.path.exists(file_path):
        print(f"ℹ File {file_path} chưa tồn tại. Khởi tạo dữ liệu rỗng.")
        return []
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            return json.load(f)
    except Exception as e:
        print(f" Lỗi khi đọc file: {e}")
        return []

def TimDanhMuc(ma_dm):
    """Tìm Danh mục theo Mã DM (Category ID). Trả về dictionary Danh mục hoặc None."""
    for dm in data:
        if dm['ma'] == ma_dm:
            return dm
    return None

def NhapSanPhamMoi():
    """Nhập thông tin sản phẩm và thêm vào danh mục tương ứng."""
    global data

    print("\n--- Nhập Danh Mục ---")
    ma_dm = input("Nhập Mã Danh Mục (DM): ")
    ten_dm = input("Nhập Tên Danh Mục: ")

    danh_muc = TimDanhMuc(ma_dm)
    if not danh_muc:
        danh_muc = {'ma': ma_dm, 'ten': ten_dm, 'san_phams': []}
        data.append(danh_muc)
        print(f" Đã tạo Danh mục mới: {ten_dm}")

    print("\n Nhập Sản Phẩm ")
    while True:
        ma_sp = input("Nhập Mã Sản Phẩm (SP): ")
        if any(sp['ma'] == ma_sp for sp in danh_muc['san_phams']):
            print("Mã SP này đã tồn tại trong danh mục. Vui lòng nhập lại.")
        else:
            break

    ten_sp = input("Nhập Tên Sản Phẩm: ")
    try:
        don_gia = float(input("Nhập Đơn Giá (Số): "))
    except ValueError:
        print(" Đơn giá không hợp lệ. Thao tác bị hủy.")
        return

    san_pham = {'ma': ma_sp, 'ten': ten_sp, 'don_gia': don_gia}
    danh_muc['san_phams'].append(san_pham)
    print(f" Đã thêm Sản phẩm '{ten_sp}' vào Danh mục '{danh_muc['ten']}'")


def XuatToanBoDuLieu():
    """Xuất toàn bộ danh sách Danh mục và Sản phẩm ra màn hình."""
    if not data:
        print("Danh sách trống.")
        return

    print("\n=============================================")
    print("           DANH SÁCH QUẢN LÝ SẢN PHẨM")
    print("=============================================")

    for dm in data:
        print(f"[{dm['ma']}] DANH MỤC: {dm['ten']}")

        if not dm['san_phams']:
            print("    -> (Không có sản phẩm nào)")
            continue

        print(f"    {'Mã SP':<8}{'Tên Sản Phẩm':<25}{'Đơn Giá':>10}")
        print("    " + "-" * 43)
        for sp in dm['san_phams']:
            print(f"    {sp['ma']:<8}{sp['ten']:<25}{sp['don_gia']:>10,.0f}")
        print("-" * 50)

def TimKiemSanPham():
    """Tìm kiếm sản phẩm theo Mã hoặc Tên."""
    keyword = input("Nhập Mã SP hoặc Tên SP cần tìm: ").lower()
    results = []

    for dm in data:
        for sp in dm['san_phams']:
            if keyword in sp['ma'].lower() or keyword in sp['ten'].lower():
                results.append((dm['ten'], sp)) # Lưu cả tên danh mục

    if results:
        print("\n--- KẾT QUẢ TÌM KIẾM ---")
        for dm_ten, sp in results:
             print(f"[DM: {dm_ten}] | Mã: {sp['ma']} | Tên: {sp['ten']} | Giá: {sp['don_gia']:,.0f}")
    else:
        print("Không tìm thấy sản phẩm nào phù hợp.")

def SapXepSanPham():
    """Sắp xếp toàn bộ sản phẩm trong tất cả các danh mục theo Đơn giá Giảm Dần."""
    all_products = []
    for dm in data:
        all_products.extend(dm['san_phams'])

    if not all_products:
        print("Không có sản phẩm nào để sắp xếp.")
        return
    all_products.sort(key=lambda x: x['don_gia'], reverse=True)

    print("\n--- SẢN PHẨM SẮP XẾP THEO ĐƠN GIÁ GIẢM DẦN (Toàn bộ) ---")
    print(f"{'Mã SP':<8}{'Tên Sản Phẩm':<25}{'Đơn Giá':>10}")
    print("-" * 43)
    for sp in all_products:
        print(f"{sp['ma']:<8}{sp['ten']:<25}{sp['don_gia']:>10,.0f}")

def Menu():
    global data
    data = DocDuLieu(FILE_NAME)

    while True:
        print("\n===== PHẦN MỀM QUẢN LÝ SẢN PHẨM =====")
        print("1. Nhập sản phẩm mới (bao gồm tạo/chọn danh mục)")
        print("2. Xuất toàn bộ danh sách")
        print("3. Tìm kiếm sản phẩm")
        print("4. Sắp xếp sản phẩm theo Đơn giá (Giảm dần)")
        print("0. Lưu và Thoát")
        print("--------------------------------------")

        choice = input("Nhập lựa chọn của bạn: ")

        if choice == '1':
            NhapSanPhamMoi()
        elif choice == '2':
            XuatToanBoDuLieu()
        elif choice == '3':
            TimKiemSanPham()
        elif choice == '4':
            SapXepSanPham()
        elif choice == '0':
            LuuDuLieu(FILE_NAME, data)
            print("Đã lưu dữ liệu và thoát chương trình. Tạm biệt!")
            break
        else:
            print("Lựa chọn không hợp lệ. Vui lòng thử lại.")
if __name__ == "__main__":
    Menu()
```

```python
#Câu 10
import json
import os
FILE_NAME = "student_manager_data.json"
data_classes = []

def LuuDuLieu(file_path, du_lieu):
    """Ghi toàn bộ cấu trúc dữ liệu Lớp học vào file JSON."""
    try:
        with open(file_path, 'w', encoding='utf-8') as f:
            json.dump(du_lieu, f, ensure_ascii=False, indent=4)
        print(f"\n Đã lưu dữ liệu thành công vào {file_path}")
    except Exception as e:
        print(f" Lỗi khi lưu file: {e}")

def DocDuLieu(file_path):
    """Đọc dữ liệu từ file JSON. Nếu file không tồn tại, trả về list rỗng."""
    if not os.path.exists(file_path):
        print(f"ℹ️ File {file_path} chưa tồn tại. Khởi tạo dữ liệu rỗng.")
        return []
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            return json.load(f)
    except Exception as e:
        print(f"❌ Lỗi khi đọc file: {e}")
        return []

def TimLopHoc(ma_lop):
    """Tìm Lớp học theo Mã lớp. Trả về dictionary Lớp học hoặc None."""
    for lop in data_classes:
        if lop['ma'] == ma_lop:
            return lop
    return None

def NhapSinhVienMoi():
    """Nhập thông tin sinh viên và thêm vào lớp học tương ứng."""
    global data_classes

    print("\n--- Nhập Lớp Học ---")
    ma_lop = input("Nhập Mã Lớp: ")
    ten_lop = input("Nhập Tên Lớp: ")

    lop_hoc = TimLopHoc(ma_lop)
    if not lop_hoc:
        lop_hoc = {'ma': ma_lop, 'ten': ten_lop, 'sinh_viens': []}
        data_classes.append(lop_hoc)
        print(f" Đã tạo Lớp học mới: {ten_lop}")

    print("\n--- Nhập Sinh Viên ---")
    while True:
        ma_sv = input("Nhập Mã Sinh Viên (SV): ")
        if any(sv['ma'] == ma_sv for sv in lop_hoc['sinh_viens']):
            print("Mã SV này đã tồn tại trong lớp. Vui lòng nhập lại.")
        else:
            break

    ten_sv = input("Nhập Tên Sinh Viên: ")
    try:
        nam_sinh = int(input("Nhập Năm Sinh (Số nguyên): "))
    except ValueError:
        print(" Năm sinh không hợp lệ. Thao tác bị hủy.")
        return

    sinh_vien = {'ma': ma_sv, 'ten': ten_sv, 'nam_sinh': nam_sinh}
    lop_hoc['sinh_viens'].append(sinh_vien)
    print(f" Đã thêm Sinh viên '{ten_sv}' vào Lớp '{lop_hoc['ten']}'")


def XuatToanBoDuLieu():
    """Xuất toàn bộ danh sách Lớp học và Sinh viên ra màn hình."""
    if not data_classes:
        print("Danh sách trống.")
        return

    print("\n=============================================")
    print("           DANH SÁCH QUẢN LÝ SINH VIÊN")
    print("=============================================")

    for lop in data_classes:
        print(f"[{lop['ma']}] LỚP HỌC: {lop['ten']}")

        if not lop['sinh_viens']:
            print("    -> (Không có sinh viên nào)")
            continue

        print(f"    {'Mã SV':<10}{'Tên Sinh Viên':<25}{'Năm Sinh':>10}")
        print("    " + "-" * 45)
        for sv in lop['sinh_viens']:
            print(f"    {sv['ma']:<10}{sv['ten']:<25}{sv['nam_sinh']:>10}")
        print("-" * 50)

def TimKiemSinhVien():
    """Tìm kiếm sinh viên theo Mã hoặc Tên."""
    keyword = input("Nhập Mã SV hoặc Tên SV cần tìm: ").lower()
    results = []

    for lop in data_classes:
        for sv in lop['sinh_viens']:
            if keyword in sv['ma'].lower() or keyword in sv['ten'].lower():
                results.append((lop['ten'], sv)) # Lưu cả tên lớp

    if results:
        print("\n--- KẾT QUẢ TÌM KIẾM ---")
        for lop_ten, sv in results:
             print(f"[Lớp: {lop_ten}] | Mã: {sv['ma']} | Tên: {sv['ten']} | Năm Sinh: {sv['nam_sinh']}")
    else:
        print("Không tìm thấy sinh viên nào phù hợp.")

def SapXepSinhVien():
    """Sắp xếp toàn bộ sinh viên trong tất cả các lớp theo Năm Sinh (Giảm dần/Tăng dần)."""
    all_students = []
    for lop in data_classes:
        all_students.extend(lop['sinh_viens'])

    if not all_students:
        print("Không có sinh viên nào để sắp xếp.")
        return
    all_students.sort(key=lambda x: x['nam_sinh'], reverse=False)

    print("\n--- SINH VIÊN SẮP XẾP THEO NĂM SINH TĂNG DẦN (Toàn bộ) ---")
    print(f"{'Mã SV':<10}{'Tên Sinh Viên':<25}{'Năm Sinh':>10}")
    print("-" * 45)
    for sv in all_students:
        print(f"{sv['ma']:<10}{sv['ten']:<25}{sv['nam_sinh']:>10}")


def Menu():
    global data_classes
    data_classes = DocDuLieu(FILE_NAME)

    while True:
        print("\n===== PHẦN MỀM QUẢN LÝ SINH VIÊN =====")
        print("1. Nhập sinh viên mới (bao gồm tạo/chọn lớp)")
        print("2. Xuất toàn bộ danh sách")
        print("3. Tìm kiếm sinh viên")
        print("4. Sắp xếp sinh viên theo Năm Sinh (Tăng dần)")
        print("0. Lưu và Thoát")
        print("--------------------------------------")

        choice = input("Nhập lựa chọn của bạn: ")

        if choice == '1':
            NhapSinhVienMoi()
        elif choice == '2':
            XuatToanBoDuLieu()
        elif choice == '3':
            TimKiemSinhVien()
        elif choice == '4':
            SapXepSinhVien()
        elif choice == '0':
            # 2. Lưu dữ liệu và thoát
            LuuDuLieu(FILE_NAME, data_classes)
            print("Đã lưu dữ liệu và thoát chương trình. Tạm biệt!")
            break
        else:
            print("Lựa chọn không hợp lệ. Vui lòng thử lại.")

# Chạy chương trình
if __name__ == "__main__":
    Menu()
```

```python
#Câu 11
import openpyxl
from openpyxl import Workbook, load_workbook
import os
FILE_NAME = "nhanvien_data.xlsx"
HEADERS = ["STT", "Mã", "Tên", "Tuổi"]
EMPLOYEE_DATA = [
    [1, 'NV1', 'An', 18],
    [2, 'NV2', 'Lành', 22],
    [3, 'NV3', 'Giải', 20],
    [4, 'NV4', 'Thoát', 19],
    [5, 'NV5', 'Hạnh', 25],
    [6, 'NV6', 'Phúc', 24],
]

def LuuNhanVienVaoFile(file_path, data_list):
    """Ghi dữ liệu Nhân viên (bao gồm cả tiêu đề) vào file Excel."""
    try:
        wb = Workbook()
        ws = wb.active
        ws.title = "DanhSachNhanVien"
        ws.append(HEADERS)
        for row_data in data_list:
            ws.append(row_data)
        ws.column_dimensions['A'].width = 5
        ws.column_dimensions['B'].width = 10
        ws.column_dimensions['C'].width = 15
        ws.column_dimensions['D'].width = 10

        wb.save(file_path)
        print(f"Đã lưu {len(data_list)} nhân viên vào file Excel: {file_path}")
    except Exception as e:
        print(f" Lỗi khi lưu file Excel: {e}")

def DocNhanVienTuFile(file_path):
    """Đọc dữ liệu Nhân viên từ file Excel. Trả về list data."""
    if not os.path.exists(file_path):
        print(f"ℹ File {file_path} chưa tồn tại.")
        return []

    try:
        wb = load_workbook(file_path)
        ws = wb.active

        data_list = []
        for i, row in enumerate(ws.values):
            if i == 0: continue
            if len(row) >= 4 and row[3] is not None:
                try:
                    tuoi = int(row[3])
                except ValueError:
                    tuoi = 0

                data_list.append([row[0], row[1], row[2], tuoi])

        return data_list
    except Exception as e:
        print(f" Lỗi khi đọc file Excel: {e}")
        return []

def XuatDanhSach(data_list):
    """Xuất danh sách Nhân viên ra màn hình."""
    if not data_list:
        print("Danh sách nhân viên trống.")
        return

    print("\n--- DANH SÁCH NHÂN VIÊN ---")
    print(f"{HEADERS[0]:<5}{HEADERS[1]:<10}{HEADERS[2]:<15}{HEADERS[3]:<5}")
    print("-" * 35)

    for nv in data_list:
        print(f"{nv[0]:<5}{nv[1]:<10}{nv[2]:<15}{nv[3]:<5}")
    print("----------------------------")

def SapXepNhanVien(data_list):
    """Sắp xếp Nhân viên theo Tuổi tăng dần."""
    if not data_list:
        print("Không có dữ liệu để sắp xếp.")
        return data_list
    data_list.sort(key=lambda x: x[3], reverse=False)

    print("\n Đã sắp xếp Nhân viên theo Tuổi (tăng dần).")
    return data_list

def Menu():
    global EMPLOYEE_DATA
    current_data = DocNhanVienTuFile(FILE_NAME)
    if not current_data:
        current_data = EMPLOYEE_DATA

    while True:
        print("\n===== PHẦN MỀM QUẢN LÝ NHÂN VIÊN (EXCEL) =====")
        print("1. Ghi dữ liệu nhân viên vào File Excel")
        print("2. Đọc và Xuất danh sách nhân viên từ File Excel")
        print("3. Sắp xếp Nhân viên theo Tuổi (Tăng dần) và Xuất")
        print("0. Thoát chương trình")
        print("-------------------------------------------------")

        choice = input("Nhập lựa chọn của bạn: ")

        if choice == '1':
            LuuNhanVienVaoFile(FILE_NAME, current_data)
        elif choice == '2':
            current_data = DocNhanVienTuFile(FILE_NAME)
            XuatDanhSach(current_data)
        elif choice == '3':
            current_data = SapXepNhanVien(current_data)
            XuatDanhSach(current_data)
        elif choice == '0':
            print("Đã thoát chương trình. Tạm biệt!")
            break
        else:
            print("Lựa chọn không hợp lệ. Vui lòng thử lại.")
if __name__ == "__main__":
    Menu()
```

```python
#Câu 12
import csv
import random
import os

# Định nghĩa tên file CSV
FILE_NAME = "random_numbers.csv"
NUM_ROWS = 10
NUM_COLS = 10

def LuuFileCSVNgauNhien(file_path):
    """
    Tạo file CSV chứa 10 dòng, mỗi dòng 10 số nguyên ngẫu nhiên (0-99),
    ngăn cách bởi dấu chấm phẩy (;).
    """
    try:
        with open(file_path, 'w', newline='', encoding='utf-8') as f:
            writer = csv.writer(f, delimiter=';')

            print(f"\n--- Dữ liệu đang ghi vào {file_path} ---")
            for i in range(NUM_ROWS):
                random_row = [random.randint(0, 99) for _ in range(NUM_COLS)]
                writer.writerow(random_row)
                print(";".join(map(str, random_row)))

        print(f"\n Đã tạo file CSV thành công: {file_path}")
    except Exception as e:
        print(f" Lỗi khi tạo file CSV: {e}")

def DocFileVaTinhTong(file_path):
    """
    Đọc file CSV, tính tổng giá trị của các phần tử trên mỗi dòng và xuất ra màn hình.
    """
    if not os.path.exists(file_path):
        print(f" Lỗi: Không tìm thấy file '{file_path}'. Vui lòng tạo file trước.")
        return

    print("\n--- Tổng giá trị các phần tử trên mỗi dòng ---")

    try:
        with open(file_path, 'r', newline='', encoding='utf-8') as f:
            reader = csv.reader(f, delimiter=';')

            for i, row in enumerate(reader):
                line_sum = 0
                for element in row:
                    try:
                        number = int(element.strip())
                        line_sum += number
                    except ValueError:
                        continue

                print(f"Dòng {i+1}: Tổng = {line_sum}")

        print("---------------------------------------------")

    except Exception as e:
        print(f" Lỗi khi đọc file CSV: {e}")

if __name__ == "__main__":
    LuuFileCSVNgauNhien(FILE_NAME)
    DocFileVaTinhTong(FILE_NAME)
```

```python
#Câu 13
import xml.etree.ElementTree as ET
import os
FILE_NHOM_THIET_BI = "nhomthietbi.xml"
FILE_THIET_BI = "thietbi.xml"

def TaoFileXMLMau():
    """Tạo hai file XML mẫu: nhomthietbi.xml và thietbi.xml"""

    # Dữ liệu File Nhóm Thiết Bị
    xml_nhom = """<?xml version="1.0" encoding="UTF-8" ?>
<nhoms>
  <nhom>
    <ma>n1</ma>
    <ten>Nhóm 1</ten>
  </nhom>
  <nhom>
    <ma>n2</ma>
    <ten>Nhóm 2</ten>
  </nhom>
  <nhom>
    <ma>n3</ma>
    <ten>Nhóm 3</ten>
  </nhom>
</nhoms>"""

    # Dữ liệu File Thiết Bị
    xml_thiet_bi = """<?xml version="1.0" encoding="UTF-8" ?>
<thietbis>
  <thietbi manhso="n1">
    <ma>tb1</ma>
    <ten>Thiết bị 1</ten>
  </thietbi>
  <thietbi manhso="n1">
    <ma>tb2</ma>
    <ten>Thiết bị 2</ten>
  </thietbi>
  <thietbi manhso="n2">
    <ma>tb3</ma>
    <ten>Thiết bị 3</ten>
  </thietbi>
  <thietbi manhso="n3">
    <ma>tb4</ma>
    <ten>Thiết bị 4</ten>
  </thietbi>
  <thietbi manhso="n3">
    <ma>tb5</ma>
    <ten>Thiết bị 5</ten>
  </thietbi>
</thietbis>"""

    try:
        with open(FILE_NHOM_THIET_BI, 'w', encoding='utf-8') as f:
            f.write(xml_nhom)
        with open(FILE_THIET_BI, 'w', encoding='utf-8') as f:
            f.write(xml_thiet_bi)
        print(f" Đã tạo thành công hai file XML mẫu.")
    except Exception as e:
        print(f" Lỗi khi tạo file XML: {e}")

def DocVaLienKetDuLieu():
    """Đọc dữ liệu từ 2 file XML và liên kết Thiết bị vào Nhóm tương ứng."""
    try:
        tree = ET.parse(FILE_NHOM_THIET_BI)
        root = tree.getroot()
        for nhom_node in root.findall('nhom'):
            ma = nhom_node.find('ma').text
            ten = nhom_node.find('ten').text
            nhom_thiet_bi[ma] = {'ma': ma, 'ten': ten, 'thiet_bis': []}
    except Exception as e:
        print(f" Lỗi khi đọc file {FILE_NHOM_THIET_BI}: {e}")
        return []
    try:
        tree = ET.parse(FILE_THIET_BI)
        root = tree.getroot()
        for tb_node in root.findall('thietbi'):
            ma_nhom = tb_node.get('manhso')
            ma = tb_node.find('ma').text
            ten = tb_node.find('ten').text

            thiet_bi = {'ma': ma, 'ten': ten}
            if ma_nhom in nhom_thiet_bi:
                nhom_thiet_bi[ma_nhom]['thiet_bis'].append(thiet_bi)
    except Exception as e:
        print(f" Lỗi khi đọc file {FILE_THIET_BI}: {e}")
        return []
    return list(nhom_thiet_bi.values())


def XuatDanhSachNhomThietBi(data_list):
    """Yêu cầu: Hiển thị danh sách Nhóm thiết bị."""
    if not data_list:
        print("Danh sách Nhóm thiết bị trống.")
        return

    print("\n===============================================")
    print("           DANH SÁCH NHÓM THIẾT BỊ")
    print("===============================================")

    for dm in data_list:
        so_luong = len(dm['thiet_bis'])
        print(f"[{dm['ma']}] NHÓM: {dm['ten']} (Số lượng TB: {so_luong})")


def XuatDanhSachThietBi(data_list):
    """Yêu cầu: Hiển thị danh sách Thiết bị."""
    if not data_list:
        print("Danh sách trống.")
        return

    print("\n===============================================")
    print("           DANH SÁCH THIẾT BỊ")
    print("===============================================")

    for dm in data_list:
        print(f"[{dm['ma']}] LỚP: {dm['ten']}")

        if not dm['thiet_bis']:
            print("    -> (Không có thiết bị nào)")
            continue

        print(f"    {'Mã TB':<8}{'Tên Thiết Bị':<25}")
        print("    " + "-" * 35)
        for tb in dm['thiet_bis']:
            print(f"    {tb['ma']:<8}{tb['ten']:<25}")
        print("-" * 50)


def LocNhomThietBiNhieuNhat(data_list):
    """Yêu cầu: Lọc Danh sách thiết bị theo số lượng thiết bị nhiều nhất."""
    if not data_list:
        print("Không có dữ liệu để lọc.")
        return
    data_list.sort(key=lambda x: len(x['thiet_bis']), reverse=True)
    max_count = len(data_list[0]['thiet_bis']) if data_list else 0

    if max_count == 0:
        print("\nKhông có thiết bị nào trong các nhóm.")
        return

    print("\n--- NHÓM THIẾT BỊ CÓ SỐ LƯỢNG TB NHIỀU NHẤT ---")
    for dm in data_list:
        if len(dm['thiet_bis']) == max_count:
            print(f"[{dm['ma']}] NHÓM: {dm['ten']} (Số lượng TB: {max_count})")

def Menu():
    TaoFileXMLMau()
    data = DocVaLienKetDuLieu()

    while True:
        print("\n===== QUẢN LÝ THIẾT BỊ (2 File XML) =====")
        print("1. Hiển thị danh sách Nhóm thiết bị")
        print("2. Hiển thị danh sách Thiết bị (theo Nhóm)")
        print("3. Lọc Nhóm thiết bị có số lượng TB nhiều nhất")
        print("0. Thoát")
        print("------------------------------------------")

        choice = input("Nhập lựa chọn của bạn: ")

        if choice == '1':
            XuatDanhSachNhomThietBi(data)
        elif choice == '2':
            XuatDanhSachThietBi(data)
        elif choice == '3':
        elif choice == '0':
            print("Đã thoát chương trình. Tạm biệt!")
            break
        else:
            print("Lựa chọn không hợp lệ. Vui lòng thử lại.")

if __name__ == "__main__":
    Menu()
```
