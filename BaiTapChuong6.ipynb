```python
#Câu 1
from random import randrange
print("Chương trình xử lý List")
n=int(input("Nhập số phần tử"))
lst=[0]*n
for i in range(n):
 lst[i]=randrange(-100,100)
print("List ngẫu nhiên là:")
print(lst)
print("Mời bạn thêm số mới:")
value=int(input())
lst.append(value)
print(lst)
print("bạn muốn đếm số nào:")
k=int(input())
dem=lst.count(k)
print(k,"xuất hiện ",dem,"trong list")
def CheckPrime(n):
  d=0
  for i in range(1,n+1):
    if n%i ==0:
      d+=1
    return d==2
demnt=0
tongnt=0
for x in lst:
 if CheckPrime(x):
  demnt+=1
  tongnt+=x
print("Có ",demnt,"số nguyên tố trong list")
print("Tổng=",tongnt)
lst.sort()
print("List sau khi sort:")
print(lst)

del lst
print("List sau khi xóa:")
print(lst)
```

```python
#Câu 3
from random import randrange
def TaoMaTran(m,n):
 D=[]
 for i in range(m):
  row=[]
 for j in range(n):
  row.append(randrange(100))
 D.append(row)
 return D
def XuatMaTran(D):
 for row in D:
  for element in row:
    print(element,end='\t')
 print()
def LayDong(r):
 R=D[r]
 return R
def XuatList1Chieu(R):
 for element in R:
  print(element,end='\t')
def LayCot(c):
 C=[]
 for i in range(len(D)):
  C.append(D[i][c])
 return C
def MAX(D):
 max=D[0][0]
 for i in range(len(D)):
  for j in range(len(D[i])):
    if(max<D[i][j]):
      max=D[i][j]
 return max
print("Nhập số dòng:")
m=int(input())
print("Nhập số cột:")
n=int(input())
D=TaoMaTran(m,n)
XuatMaTran(D)
print("Mời bạn nhập dòng muốn xuất:")
r=int(input())
XuatList1Chieu(LayDong(r))
print("Mời bạn nhập cột muốn xuất:")
c=int(input())
XuatList1Chieu(LayCot(c))
max=MAX(D)
print("Số lớn nhất trong ma trận=",max)
```

```python
#Câu 6
import random

def generate_unique_random_list():
    """Nhập N và tạo list N số ngẫu nhiên không trùng nhau."""

    try:
        N = int(input("Nhập số lượng phần tử N cần tạo: "))
        if N <= 0:
            print(" N phải là số nguyên dương.")
            return
        MAX_RANGE = 2 * N + 10
        random_set = set()

        while len(random_set) < N:
            new_number = random.randint(1, MAX_RANGE)
            random_set.add(new_number)
        result_list = list(random_set)

        print(f"\n List {N} số ngẫu nhiên KHÔNG TRÙNG NHAU (Phạm vi 1-{MAX_RANGE}):")
        print(result_list)

    except ValueError:
        print(" Lỗi nhập liệu: Vui lòng nhập một số nguyên hợp lệ.")

generate_unique_random_list()
```

```python
#Câu 7
def input_increasing_sequence():
    print("\n Nhập dãy số tăng dần ")
    print("Nhập 'q' hoặc nhấn Enter để kết thúc nhập liệu.")

    number_list = []

    while True:
        try:
            user_input = input(f"Nhập số tiếp theo (Hiện tại: {number_list}): ").strip()
            if user_input.lower() == 'q' or user_input == '':
                break
            new_number = float(user_input)
            if number_list:
                last_number = number_list[-1]
                if new_number <= last_number:
                    print(f"**Nhập sai quy cách!** Số {new_number} phải lớn hơn số cuối cùng ({last_number}). Vui lòng nhập lại.")
                    continue
            number_list.append(new_number)

        except ValueError:
            print(" Lỗi nhập liệu: Vui lòng nhập một giá trị số hợp lệ.")
    print("\n--- KẾT QUẢ ---")
    if number_list:
        print(f" Dãy số tăng dần đã nhập là: **{number_list}**")
    else:
        print("Không có số nào được nhập.")

# Chạy hàm
input_increasing_sequence()
```

```python
#Câu 8
def sort_real_numbers_descending():
    number_list = []
    try:
        N = int(input("Nhập số lượng phần tử N (nguyên dương): "))
        if N <= 0:
            print("❌ Số lượng phần tử N phải là số nguyên dương.")
            return
        print(f"Bắt đầu nhập {N} số thực:")
        for i in range(N):
            while True:
                try:
                    num = float(input(f"  Nhập phần tử M[{i + 1}/{N}] (số thực): "))
                    number_list.append(num)
                    break
                except ValueError:
                    print(" Lỗi: Vui lòng nhập một số thực hợp lệ.")
    except ValueError:
        print(" Lỗi nhập liệu: Vui lòng nhập số nguyên dương hợp lệ cho N.")
        return
    number_list.sort(reverse=True)
    print("\n--- KẾT QUẢ SẮP XẾP ---")
    print(f"Dãy số thực đã nhập: {number_list}")
    print(f" Dãy số sau khi sắp xếp **GIẢM DẦN** là: **{number_list}**")
sort_real_numbers_descending()
```

```python
#Câu 9
print("Nhập các số tự nhiên, cách nhau bằng dấu phẩy hoặc dấu cách:")
input_str = input()
try:
    if ',' in input_str:
        M = [int(s.strip()) for s in input_str.split(',') if s.strip().isdigit()]
    else:
        M = [int(s.strip()) for s in input_str.split() if s.strip().isdigit()]
except ValueError:
    print("Lỗi: Đầu vào không hợp lệ. Vui lòng chỉ nhập các số tự nhiên.")
    M = []
def is_prime(n):
    """Kiểm tra xem một số có phải là số nguyên tố hay không."""
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
if M:
    so_le = []
    so_chan = []
    so_nguyen_to = []
    khong_phai_nguyen_to = []

    for num in M:
        if num % 2 != 0:
            so_le.append(num)
        else:
            so_chan.append(num)
        if is_prime(num):
            so_nguyen_to.append(num)
        else:
            khong_phai_nguyen_to.append(num)
    print("\n--- KẾT QUẢ XỬ LÝ MẢNG ---")
    print(f"Mảng đã nhập: {M}")
    print("-" * 30)
    print(f"Dòng 1: Gồm các số lẻ: {so_le} -> Tổng cộng có {len(so_le)} số lẻ.")
    print(f"Dòng 2: Gồm các số chẵn: {so_chan} -> Tổng cộng có {len(so_chan)} số chẵn.")
    print(f"Dòng 3: Gồm các số nguyên tố: {so_nguyen_to}.")
    print(f"Dòng 4: Gồm các số không phải là số nguyên tố: {khong_phai_nguyen_to}.")
else:
    print("Không có dữ liệu hợp lệ để xử lý.")
```

```python
#Câu 10
import numpy as np
def nhap_ma_tran(ten_ma_tran):
    print(f"\n--- NHẬP MA TRẬN {ten_ma_tran} ---")
    while True:
        try:
            hang = int(input(f"Nhập số hàng của ma trận {ten_ma_tran}: "))
            cot = int(input(f"Nhập số cột của ma trận {ten_ma_tran}: "))
            if hang <= 0 or cot <= 0:
                print("Số hàng và số cột phải là số nguyên dương. Vui lòng nhập lại.")
                continue
            break
        except ValueError:
            print("Đầu vào không hợp lệ. Vui lòng nhập một số nguyên.")
    phan_tu = []
    print(f"Nhập các phần tử cho ma trận {ten_ma_tran} ({hang}x{cot}) theo thứ tự hàng-cột:")
    for i in range(hang):
        while True:
            try:
                row_input = input(f"Nhập {cot} phần tử của hàng {i+1} (cách nhau bằng dấu cách): ")
                row_elements = [float(s) for s in row_input.split()]
                if len(row_elements) != cot:
                    print(f"Bạn đã nhập {len(row_elements)} phần tử. Vui lòng nhập đủ {cot} phần tử cho hàng này.")
                    continue
                phan_tu.extend(row_elements)
                break
            except ValueError:
                print("Đầu vào không hợp lệ. Vui lòng nhập các giá trị số.")
    matrix = np.array(phan_tu).reshape((hang, cot))
    return matrix
def cong_ma_tran(A, B):
    print("\n--- CỘNG 2 MA TRẬN A VÀ B ---")
    if A.shape != B.shape:
        print("LỖI: Hai ma trận không cùng kích thước. Không thể cộng.")
        return None
    ket_qua = A + B
    print(f"Ma trận Kết quả (A + B) là:\n{ket_qua}")
    return ket_qua
def tinh_ma_tran_chuyen_vi(M):
    print("\n--- TÍNH MA TRẬN CHUYỂN VỊ (HOÁN VỊ) ---")
    chuyen_vi = M.T
    print(f"Ma trận gốc:\n{M}")
    print(f"Ma trận Chuyển vị M^T:\n{chuyen_vi}")
    return chuyen_vi
if __name__ == "__main__":
    A = nhap_ma_tran("A")
    B = nhap_ma_tran("B")
    print("\n--- MA TRẬN ĐÃ NHẬP ---")
    print(f"Ma trận A:\n{A}")
    print(f"Ma trận B:\n{B}")
    C = cong_ma_tran(A, B)
    print("\n--- ÁP DỤNG HÀM CHUYỂN VỊ ---")
    A_chuyen_vi = tinh_ma_tran_chuyen_vi(A)
    print("-" * 30)
    B_chuyen_vi = tinh_ma_tran_chuyen_vi(B)
    print("\nChương trình xử lý ma trận đã hoàn tất.")
```
